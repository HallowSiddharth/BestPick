```python
import pandas as pd

def transfer_columns(input_csv, output_csv, columns_to_transfer):
    # Read the input CSV file
    df = pd.read_csv(input_csv)

    # Select the specified columns
    selected_columns = df[columns_to_transfer]

    # Write the selected columns to a new CSV file
    selected_columns.to_csv(output_csv, index=False)

# Example usage:
input_csv_file = 'Reviews.csv'  # Replace with your input CSV file
output_csv_file = 'dataset.csv'  # Replace with your desired output CSV file
columns_to_transfer = ['Score', 'Text']  # Replace with the column names you want to transfer

transfer_columns(input_csv_file, output_csv_file, columns_to_transfer)
```
